```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies_reviews = pd.read_csv("../data/movies_collection.csv")
movies = pd.read_csv("../data/movies.csv")
reviews = pd.read_csv("../data/reviews.csv")
```

```python
def plot_missing(df):
    missing_values = df.isnull().sum()

    print((missing_values.sum() / (df.shape[0] * df.shape[1])) * 100)

    plt.figure(figsize=(10, 6))
    missing_values.plot(kind='bar', color='skyblue')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.title('Missing Values per Column')
    plt.show()
```

```python
plot_missing(movies)
```

```python
plot_missing(reviews)
```

```python
plot_missing(movies_reviews)
```

```python
def plot_duplicates(df):
    
    print(df.duplicated().sum())
    
```

```python
plot_duplicates(movies)
```

```python
plot_duplicates(reviews)
```

```python
plot_duplicates(movies_reviews)
```
